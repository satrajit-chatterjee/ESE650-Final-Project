```python
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path

# Read the data
data = np.genfromtxt("state_observation.csv", delimiter=",")

data = data[1:,:]
velocity_pred = data[:,0]
velocity_actual = data[:,-1]
vel_x = data[:,1]
vel_y = data[:,2]
slip_angle = data[:,3]
print(velocity_pred)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(velocity_pred, label="Predicted Velocity")
plt.plot(velocity_actual, label="Actual Velocity")
plt.xlabel("Data Point")
plt.ylabel("Velocity")
plt.title("Comparison of Predicted and Actual Velocity")
plt.legend()
plt.grid()
plt.show()
```
